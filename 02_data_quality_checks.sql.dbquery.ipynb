```sql
-- =========================================================
-- File: 03_data_quality_checks.sql
-- Purpose:
-- Perform data quality validation checks
-- on Silver layer transaction data.
--
-- Checks include:
-- - Null validation
-- - Amount validation
-- - Invalid numeric values
-- - Duplicate detection
-- =========================================================


-- =========================================================
-- QC Check: Null transaction IDs
-- =========================================================

SELECT *
FROM transactions.silver.transactions
WHERE transaction_id IS NULL;


-- =========================================================
-- QC Check: Validate total amount calculation
-- =========================================================

SELECT *
FROM transactions.silver.transactions
WHERE ROUND(total_amount, 2)
    != ROUND(quantity * unit_price, 2);


-- =========================================================
-- QC Check: Invalid quantities
-- =========================================================

SELECT *
FROM transactions.silver.transactions
WHERE quantity <= 0;


-- =========================================================
-- QC Check: Invalid pricing values
-- =========================================================

SELECT *
FROM transactions.silver.transactions
WHERE unit_price <= 0;


-- =========================================================
-- QC Check: Duplicate transaction IDs
-- =========================================================

SELECT
    transaction_id,
    COUNT(*) AS duplicate_count
FROM transactions.silver.transactions
GROUP BY transaction_id
HAVING COUNT(*) > 1;
```
