```sql
-- =========================================================
-- File: 02_silver_to_gold.sql
-- Purpose:
-- Create business-ready Gold layer tables
-- from cleaned Silver layer transactional data.
--
-- Gold tables are optimized for reporting,
-- analytics, and dashboard consumption.
-- =========================================================


-- =========================================================
-- Gold Table: sales_by_store
-- Purpose:
-- Generate sales KPIs grouped by store location.
-- =========================================================

CREATE OR REPLACE TABLE transactions.gold.sales_by_store AS

SELECT
    store_location,
    COUNT(*) AS total_transactions,
    SUM(total_amount) AS total_sales,
    AVG(total_amount) AS avg_order_value

FROM transactions.silver.transactions

GROUP BY store_location;


-- =========================================================
-- Gold Table: product_sales
-- Purpose:
-- Analyze product-level sales performance.
-- =========================================================

CREATE OR REPLACE TABLE transactions.gold.product_sales AS

SELECT
    product_name,
    SUM(quantity) AS total_quantity_sold,
    SUM(total_amount) AS total_revenue

FROM transactions.silver.transactions

GROUP BY product_name

ORDER BY total_revenue DESC;
```
