```sql
-- =========================================================
-- File: 01_bronze_to_silver.sql
-- Purpose:
-- Transform raw transaction data from Bronze layer
-- into cleaned and standardized Silver layer.
--
-- Key Transformations:
-- 1. Convert timestamp columns into proper datatypes
-- 2. Standardize text formatting
-- 3. Remove extra spaces from product names
-- 4. Cast numeric columns into proper datatypes
-- 5. Prepare validated transactional dataset
-- =========================================================

CREATE OR REPLACE TABLE transactions.silver.transactions AS

SELECT

    -- Unique transaction identifier
    transaction_id,

    -- Convert raw timestamp string into timestamp datatype
    TO_TIMESTAMP(transaction_date) AS transaction_timestamp,

    -- Extract only date portion for analytics/reporting
    TO_DATE(transaction_date) AS transaction_date,

    -- Customer identifier
    customer_id,

    -- Product identifier
    product_id,

    -- Standardize product names:
    -- 1. Remove extra spaces
    -- 2. Trim leading/trailing spaces
    -- 3. Convert to proper title case
    INITCAP(
        TRIM(
            REGEXP_REPLACE(product_name, '\\s+', ' ')
        )
    ) AS product_name,

    -- Standardize category names
    INITCAP(TRIM(category)) AS category,

    -- Convert quantity into integer datatype
    CAST(quantity AS INT) AS quantity,

    -- Convert unit price into decimal datatype
    CAST(unit_price AS DECIMAL(10,2)) AS unit_price,

    -- Convert total amount into decimal datatype
    CAST(total_amount AS DECIMAL(10,2)) AS total_amount,

    -- Standardize store location names
    INITCAP(TRIM(store_location)) AS store_location,

    -- Standardize payment method names
    INITCAP(TRIM(payment_method)) AS payment_method

FROM transactions.bronze.transactions;
```
